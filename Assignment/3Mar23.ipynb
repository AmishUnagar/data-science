Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Seaborn is a powerful Python library for data visualization built on top of Matplotlib. It offers a wide range of statistical plots that are both visually appealing and informative. Here are five commonly used plots in Seaborn along with their use cases:

1. Scatter Plot:
 Scatter plots are useful for visualizing the relationship between two continuous variables. They help identify patterns, correlations, clusters, or outliers in the data.

2. Line Plot:
Line plots are used to visualize trends and changes over time. They are effective in showing the progression of data points, such as stock prices, temperature variations, or any other time series data.

3. Bar Plot:
Bar plots are suitable for comparing categorical variables. They display the distribution or relationship between a categorical variable and a continuous variable, making it easy to compare values across categories.

4. Histogram:
 Histograms are used to represent the distribution of a single variable. They show the frequency or count of data points falling into different bins or intervals, allowing you to understand the underlying distribution.

5. Box Plot:
Box plots provide a concise summary of the distribution of a continuous variable. They display the quartiles, median, and any potential outliers in the data, making them effective for identifying data skewness, variability, and comparing distributions across different groups.

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "fmri" dataset from Seaborn
fmri_df = sns.load_dataset("fmri")

# Plot a line plot using Seaborn and Matplotlib
sns.lineplot(data=fmri_df, x="timepoint", y="signal", hue="event", style="region")

# Add labels and title
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.title("FMRI Dataset Line Plot")

# Display the plot
plt.show()
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "titanic" dataset from Seaborn
titanic_df = sns.load_dataset("titanic")

# Create two subplots for the box plots
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# Plot the first box plot: age vs. pclass
sns.boxplot(ax=axes[0], x="pclass", y="age", data=titanic_df)
axes[0].set_xlabel("Passenger Class")
axes[0].set_ylabel("Age")
axes[0].set_title("Age Distribution by Passenger Class")

# Plot the second box plot: fare vs. pclass
sns.boxplot(ax=axes[1], x="pclass", y="fare", data=titanic_df)
axes[1].set_xlabel("Passenger Class")
axes[1].set_ylabel("Fare")
axes[1].set_title("Fare Distribution by Passenger Class")

# Adjust spacing between subplots
plt.tight_layout()

# Display the plots
plt.show()
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "diamonds" dataset from Seaborn
diamonds_df = sns.load_dataset("diamonds")

# Plot a histogram with hue for the 'cut' column
sns.histplot(data=diamonds_df, x="price", hue="cut", multiple="stack")

# Add labels and title
plt.xlabel("Price")
plt.ylabel("Count")
plt.title("Diamonds Dataset - Histogram of Price with Cut")

# Display the plot
plt.show()
```

Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.

```python
import seaborn as sns

# Load the "iris" dataset from Seaborn
iris_df = sns.load_dataset("iris")

# Plot a pair plot with hue for the "species" column
sns.pairplot(data=iris_df, hue="species")

# Display the plot
plt.show()
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "flights" dataset from Seaborn
flights_df = sns.load_dataset("flights").pivot("month", "year", "passengers")

# Plot a heatmap
sns.heatmap(data=flights_df, cmap="YlGnBu")

# Add labels and title
plt.xlabel("Year")
plt.ylabel("Month")
plt.title("Flights Dataset - Heatmap")

# Display the plot
plt.show()
```
