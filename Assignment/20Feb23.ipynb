Q1. Explain GET and POST methods.

GET and POST are HTTP methods used to send and receive data between a client and a server. They are two of the most commonly used methods in web development.

GET:

It is used to request a resource from a server.
It sends data in the form of a URL-encoded query string in the URL itself.
GET requests can be bookmarked and shared easily as they can be represented by a simple URL.
It is suitable for retrieving data and should not be used for sending sensitive or confidential data as the data is visible in the URL.
It is a safe and idempotent operation, meaning that multiple identical requests will have the same effect as a single request.

POST:


It is used to submit data to a server to be processed and stored.
It sends data in the form of a request body and is not visible in the URL.
POST requests cannot be bookmarked or shared as easily as GET requests.
It is suitable for sending sensitive or confidential data as the data is not visible in the URL.
It is not a safe or idempotent operation, meaning that multiple identical requests may have different effects.

Q2. Why is request used in Flask?

In Flask, a request object is used to handle incoming HTTP requests made by the client. The request object contains all the necessary information about the incoming request such as the headers, cookies, form data, URL parameters, and HTTP method type.

The request object is an instance of the Request class, which is created by Flask automatically when a request is received. It provides a way for the developer to access and manipulate the data from the request in order to generate an appropriate response.

Using the request object, you can retrieve the data from the incoming request and use it to perform various tasks like processing user input, validating user input, authenticating requests, and many more.

For example, you can use the request object to access form data submitted by the user in a POST request, like so:

```python
from flask import Flask, request

app = Flask(__name__)

@app.route('/login', methods=['POST'])
def login():
    username = request.form['username']
    password = request.form['password']
    # do something with username and password
    return 'Logged in successfully'
```

In the above example, the request object is used to retrieve the username and password submitted in the form data of the POST request. These values can then be used to authenticate the user and return an appropriate response.

Overall, the request object is a crucial part of Flask and is used extensively in web development to handle incoming HTTP requests and generate appropriate responses.

Q3. Why is redirect() used in Flask?

The redirect() function is used in Flask to redirect the user to a different URL. It is commonly used after a form has been submitted or after some processing has been completed on the server-side and the user needs to be directed to a different page.

The redirect() function takes a URL as an argument and returns a response with a status code of 302, which tells the browser to go to the new URL. The new URL can be a route within the same Flask application or it can be an external URL.

Q4. What are templates in Flask? Why is the render_template() function used?

In Flask, templates are files that contain the structure and layout of the web pages that are dynamically generated by the web application. Templates are written using a templating language, such as Jinja2, which allows developers to write HTML code with placeholders for dynamic content, such as variables or expressions.

The render_template() function in Flask is used to render a template with the dynamic content. This function takes the name of the template file as its argument and returns the rendered HTML code. The function searches for the template file in the templates directory by default, which should be located in the root directory of the Flask application.

The render_template() function can also take additional arguments, which are used to pass dynamic content to the template. For example, a view function in Flask can pass a list of blog posts to a template, which can then be rendered using a loop in the template.

```python
from flask import Flask,request,jsonify

app = Flask(__name__)

@app.route("/postman_data",methods=['POST'])
def math_operation():
    if(request.method=="POST"):
        ops =request.json['operation']
        num1= int(request.json['num1'])
        num2= int(request.json['num2'])

        if(ops=="add"):
            r= num1+num2
            result= f"sum of {str(num1)} and {str(num2)} is {str(r)}"
        elif(ops=="subtract"):
            r= num1-num2
            result= f"subtract of {str(num1)} and {str(num2)} is {str(r)}"
        elif(ops=="multiply"):
            r= num1*num2
            result= f"multiply of {str(num1)} and {str(num2)} is {str(r)}"
        elif(ops=="divide"):
            r= num1/num2
            result= f"divide of {str(num1)} and {str(num2)} is {str(r)}"

        return jsonify(result)
        

if __name__=="__main__":
    app.run(debug=True,host="0.0.0.0")
    
```
