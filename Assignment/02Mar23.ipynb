Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Matplotlib is a popular Python library used for creating visualizations, such as plots, charts, and graphs. It provides a flexible and comprehensive set of tools for generating high-quality figures.

Matplotlib is used for various purposes, including data exploration, data analysis, and data communication. It allows users to represent their data in a visually appealing and informative manner, making it easier to understand patterns, trends, and relationships within the data.

The Pyplot module in Matplotlib offers a simple and intuitive interface for creating various types of plots. Five common plots that can be created using Pyplot are:

Line plot: Used to display the relationship between two continuous variables over a continuous interval.

Scatter plot: Represents the correlation or relationship between two variables by displaying individual data points.

Bar plot: Displays categorical data using rectangular bars of different heights or lengths to represent the values.

Histogram: Shows the distribution of a single variable by dividing it into bins and displaying the frequency or count of observations in each bin.

Pie chart: Presents categorical data as slices of a pie, where each slice represents a proportion or percentage of the whole.

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

 A scatter plot is a type of plot that displays the correlation or relationship between two variables by representing individual data points on a two-dimensional plane. Each data point is plotted based on its corresponding values for the two variables, with one variable represented on the x-axis and the other on the y-axis.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y,c='r')
plt.xlabel("this is x axis")
plt.ylabel("this is y axis")
plt.title("scatter plot")
plt.show()
```

3) Why is the subplot() function used? Draw four line plots using the subplot() function.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])


plt.subplot(2,2,1)
plt.plot(x,y1)
plt.title("first")

plt.subplot(2,2,2)
plt.plot(x,y2)
plt.title("second")

plt.subplot(2,2,3)
plt.plot(x,y3)
plt.title("therd")

plt.subplot(2,2,4)
plt.plot(x,y4)
plt.title("fourth")


plt.tight_layout()

```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

A bar plot, also known as a bar chart, is a type of plot that represents categorical data using rectangular bars. The length or height of each bar corresponds to the value or frequency of the category it represents. Bar plots are commonly used to compare and display data across different categories or groups.

Bar plots are used for several purposes:

Comparison: Bar plots allow for easy visual comparison of values or frequencies between different categories or groups.

Distribution: Bar plots can show the distribution of a categorical variable by displaying the frequency or count of each category.

Patterns and trends: Bar plots help identify patterns and trends in the data, such as identifying the most frequent or least frequent categories.

Relationships: Bar plots can be used to visualize relationships between categories or groups, such as comparing data across different time periods or demographic groups.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company,profit)
plt.xlabel("company")
plt.ylabel("profile")
plt.title("bar chart for company profit")
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It displays the summary statistics, such as the minimum, first quartile (25th percentile), median (second quartile or 50th percentile), third quartile (75th percentile), and maximum, along with any outliers or extreme values.

A box plot is used for several purposes:

Descriptive statistics: Box plots provide a concise summary of the central tendency, spread, and skewness of the data.

Comparison: Box plots allow for easy visual comparison of multiple datasets or groups, enabling quick identification of differences in their distributions.

Outlier detection: Box plots can help identify outliers or extreme values that fall outside the typical range of the data.

Skewness and symmetry: Box plots provide insights into the skewness and symmetry of the data distribution based on the position of the median and the lengths of the whiskers.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

plt.boxplot([box1,box2])
```
